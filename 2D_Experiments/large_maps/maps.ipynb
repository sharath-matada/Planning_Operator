```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
maps = np.load('../dataset/street_maps_512.npy')
```

```python
maps.shape
```

```python
# Plot all 30 maps with their respective index numbers
fig, axes = plt.subplots(6, 5, figsize=(20, 24))  # 6 rows, 5 columns for 30 maps

for i, ax in enumerate(axes.flat):
    ax.imshow(maps[i], cmap='gray')
    ax.set_title(f"Map {i+1}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
